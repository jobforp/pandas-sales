```python
s_df = pd.merge(sales,products, how='outer')
s_df.head(10)
```

so the trick here is the how the merge happens. If you are familiar with SQL it should straight forward.

+ left - Use keys from left frame only
+ right -Use keys from right frame only
+ outer - Use union of keys from both frames
+ inner - Use intersection of keys from both frames

<http://pandas.pydata.org/pandas-docs/stable/merging.html#database-style-dataframe-joining-merging>
    
If you aren't familiar with these - I recommend trying each and seeing the outcome.

```python
s_df.unitsSold.count()
```

I count to make sure we didn't drop any rows due to the join

ok lets calculate gross profit margins

```python
s_df['totalsales'] = s_df.unitsSold * s_df.stdPrice
s_df['cost'] = s_df.unitsSold * s_df.stdCost
s_df['gross'] = s_df.totalsales - s_df.cost
s_df.head()
```
